```python
# Enable gradient tracking by default
x = torch.randn(5, 5, requires_grad=True)
y = torch.randn(5, 5, requires_grad=True)

print(f"x.requires_grad: {x.requires_grad}") # Output: x.requires_grad: True

# Operations inside this block will not track gradients
with torch.no_grad():
    z = x + y
    w = z * z
    print(f"z.requires_grad: {z.requires_grad}") # Output: z.requires_grad: False
    print(f"w.requires_grad: {w.requires_grad}")
```
